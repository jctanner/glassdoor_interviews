```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from make_interview_stats import get_interviews_records
from make_interview_stats import records_to_frame
```

```python
records = get_interviews_records()
(pkeys,df_records,df) = records_to_frame(records)
pkeys = sorted(pkeys)
#print(pkeys)
values = []
for dfr in df_records:
    idx = pkeys.index(dfr[0])
    values.append((idx, dfr[1]))
ndf = pd.DataFrame.from_records(values)

ndf.columns = ['method', 'percent']
#ndf.set_index('method', inplace=True)
ndf.head()
    
ax = ndf.plot(kind='scatter', x='method', y='percent', alpha=.2)
ax.set_xticklabels([''] + pkeys, rotation=90)
```
